```python
#Arithmetic Operators

print(5+6)
print(5-6)
print(5*6)
print(5/2)
print(5//2)   #Integer didision
print(5%6)
print(5**2)   #power
```

```python
#Relational Operators

print(4>5)
print(4<5)
print(4<=5)
print(4>=5)
print(4==5)
print(4!=5)
```

```python
# Logical Operators
print(1 and 0)  #dono me 1 hoga tab output me 1 hoga
print(1 or 0)   #kisime bhi 1 hoga to wo 1 dega
print(not 0)
```

```python
#Bitwise Operators

#bitwise and
print(2&3)
#bitwise or
print(2|3)
#bitwise xor
print(2^3)
#bitwise nor
print(~3)
#bitwise left shift
print(4>>2)
#bitwise right shift
print(4<<2)
```

```python
#Assignment Operators

# =
# a = 2
a=2
#a=a+2
a+=2
#a=a%2
a%=2
# a++ ++a  is not allowed in python
print(a)
```

```python
#Membership Operator

#in
#not in

print('D' in 'Delhi')
print('D' not in 'Delhi')
print(1 in [2,3,4,5,6])
```

```python
#menu driven calculator
num1=int(input('Enter the first number'))
num2=int(input('Enter the second number'))

op=input('Enter the operation:')

if op=='+':
  print(num1+num2)
elif op=='-':
  print(num1-num2)
elif op=='*':
  print(num1*num2)
elif op=='/':
  print(num1/num2)
else:
  print("Invalid operator")
```
